### importing liblaries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

### Visualizing the Data using Matplotlib/Seaborn

```python

# Generate synthetic data: Size of house (in square feet) and price (in dollars)

np.random.seed(0)
x = np.random.rand(100, 1) * 2000  # House sizes ranging from 0 to 2000 square feet
y = 50 * x + 50000 + np.random.randn(100, 1) * 10000  # Price with some random noise

# Convert the data to a pandas DataFrame for easier handling
data = pd.DataFrame({'Size': x.flatten(), 'Price': y.flatten()})

# Plot the data using Seaborn
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Size', y='Price', data=data, color='blue', label='Data points')

# Fit a linear regression model using scikit-learn
model = LinearRegression()
model.fit(x, y)

# Get the slope and intercept
slope = model.coef_[0]
intercept = model.intercept_

# Predict the price based on the house size
y_pred = model.predict(x)

# Plot the regression line
plt.plot(x, y_pred, color='red', label=f'Regression line: y = {slope[0]:.2f}x + {intercept[0]:.2f}')

# Adding labels and title
plt.title('House Price vs Size (Linear Regression)', fontsize=16)
plt.xlabel('Size (Square Feet)', fontsize=12)
plt.ylabel('Price (Dollars)', fontsize=12)
plt.legend()
plt.show()

# Print the model's slope and intercept
print(f"Slope (m): {slope[0]}")
print(f"Intercept (b): {intercept[0]}")
```

### Generate or load the dataset

```python
# Generating some sample data (for illustration)
np.random.seed(0)
X = np.random.rand(100, 1) * 10  # Independent variable (size of the house in square feet)
y = 5 * X + 3 + np.random.randn(100, 1) * 2  # Dependent variable (house price), with noise

# Visualize the generated data
plt.scatter(X, y)
plt.title('House Price vs Size (Synthetic Data)')
plt.xlabel('Size (in square feet)')
plt.ylabel('Price (in dollars)')
plt.show()
```

### Create the model and fit it

```python
# Initialize the Linear Regression model
model = LinearRegression()

# Fit the model using the data
model.fit(X, y)

# Get the slope (m) and intercept (b)
slope = model.coef_[0]
intercept = model.intercept_

print(f"Model Slope (m): {slope[0]}")
print(f"Model Intercept (b): {intercept[0]}")
```

###  Make Predictions and Visualize the Results

```python
# Make predictions using the trained model
y_pred = model.predict(X)

# Visualize the regression line along with the original data points
plt.scatter(X, y, color='blue', label='Data points')  # Original data
plt.plot(X, y_pred, color='red', label='Regression line')  # Fitted line
plt.title('Simple Linear Regression: House Price vs Size')
plt.xlabel('Size (in square feet)')
plt.ylabel('Price (in dollars)')
plt.legend()
plt.show()
```
